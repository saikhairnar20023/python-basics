```python
missing_value_per = df.isnull().sum()/df.shape[0] * 100
missing_value_per
```

```python
missing_value_grt20 = missing_value_per[missing_value_per > 20 ].keys()
missing_value_grt20
```

```python
missing_value_del = df.drop(columns = missing_value_grt20)
missing_value_del
```

```python
missing_value_del.shape
```

```python
df3_num = missing_value_del.select_dtypes(include=['int64','float64'])
df3_num
```
